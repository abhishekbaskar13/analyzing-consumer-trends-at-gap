```python
%%time
def request_page(url):
    options = Options()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument(f'user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36')
    driver = webdriver.Chrome(options=options)  
    try:
        driver.get(url)
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, 'body')))
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(random.uniform(2, 4))
        html = driver.page_source
        return html
    except Exception as e:
        print(f"An error occurred: {e}")
        return None
    finally:
        driver.quit()


product_review_list_gap = []
product_id_list_gap = []
# Example: assuming product_links_gap, product_review_list_gap, and product_id_list_gap are already defined

i = 0
for link_gap in product_links_gap:
    try:
        # Request the page HTML
        product_html_gap = request_page(link_gap)  # Assuming request_page() handles HTTP requests
        
        # Parse the HTML using BeautifulSoup
        soup_gap = bs(product_html_gap, 'html.parser')
        
        # Find the product review or description (improve this by handling None cases)
        relevant_gap = soup_gap.find('p', class_='pr-rd-description-text')
        
        if relevant_gap is not None:
            product_review_gap = relevant_gap.text.strip()  # Clean text
        else:
            product_review_gap = "No review found"
        
        # Extract product ID from the URL
        pid_gap = link_gap.split('&vid')[0].split('pid=')[-1].split('&tdp=cm_app')[0]
        
        # Append the extracted data to the respective lists
        product_review_list_gap.append(product_review_gap)
        product_id_list_gap.append(pid_gap)
        
        # Pause to prevent overwhelming the server
        time.sleep(random.uniform(3, 7))
        
        # Print progress
        i += 1
        print(f"Processed {i} products")
    
    except Exception as e:
        print(f"Error processing {link_gap}: {e}")
        continue  # Skip to the next product in case of an error
```
