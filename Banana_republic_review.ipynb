```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, ElementClickInterceptedException
from selenium.webdriver.chrome.options import Options
import time
import random

# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--window-size=1920x1080")

# Initialize the WebDriver
driver = webdriver.Chrome(options=chrome_options)

# List of product URLs
product_urls = product_links_gap

product_review_list_gap = []
product_id_list_gap = []

def click_element(driver, by, value, timeout=10):
    try:
        element = WebDriverWait(driver, timeout).until(
            EC.element_to_be_clickable((by, value))
        )
        driver.execute_script("arguments[0].click();", element)
    except (TimeoutException, ElementClickInterceptedException):
        print(f"Failed to click element: {value}")
        return False
    return True

for url in product_urls:
    try:
        driver.get(url)
        time.sleep(5)  # Wait for initial page load
        
        # Extract product ID from the URL
        product_id = url.split('pid=')[1].split('&')[0]
        product_id_list_gap.append(product_id)
        
        # Click the reviews tab
        if not click_element(driver, By.ID, "product-information-tabs-button--3"):
            raise Exception("Couldn't click reviews tab")
        
        time.sleep(2)  # Wait for tab content to load
        
        # Click the "See Reviews" button
        if not click_element(driver, By.CLASS_NAME, "pdp-review-ratings-btn"):
            raise Exception("Couldn't click 'See Reviews' button")
        
        time.sleep(2)  # Wait for reviews to load
        
        # Try to find the review text
        try:
            review_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CLASS_NAME, "pr-rd-description-text"))
            )
            review_text = review_element.text.strip()
        except (TimeoutException, NoSuchElementException):
            review_text = "No review found"
        
        product_review_list_gap.append(review_text)
        
        print(f"Processed product ID: {product_id}")
        
        # Add a random delay between requests
        time.sleep(random.uniform(3, 7))
        
    except Exception as e:
        print(f"Error processing {url}: {e}")
        product_review_list_gap.append("Error occurred")
    
    # Print progress
    print(f"Processed {len(product_review_list_gap)} out of {len(product_urls)} products")

# Close the browser
driver.quit()

# Print the results
for pid, review in zip(product_id_list_gap, product_review_list_gap):
    print(f"Product ID: {pid}")
    print(f"Review: {review}")
    print("---")
```
