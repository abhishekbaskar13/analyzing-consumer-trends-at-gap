```python
# Importing Necessary Libraries & Initialising Driver

from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

chrome_options = Options()
chrome_options.add_argument('--headless') #uncomment this if you want the browser to run queitly, i.e., it will run in the background
chrome_options.add_argument('--no-sandbox') # operate at the highest authority
chrome_options.add_argument('--disable-dev-shm-usage') #increase the RAM of chrome to load the page
```

```python
# Code to get Trending Singers in the US from the Billboards Top 100 Charts

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)


artist_list = []


# Open Billboard Hot 100 page
url = "https://www.billboard.com/charts/hot-100/"
driver.get(url)

# Wait for the page to load
time.sleep(3)  # You can adjust this based on your internet speed

# Find all artist name elements on the page
artist_elements = driver.find_elements(By.CLASS_NAME, "c-label.a-no-trucate")

# Extract the text (artist names) from the elements
artists = [element.text for element in artist_elements]

# Print the top 10 artists
print("Trending Billboard Artists (Hot 100):")
for i, artist in enumerate(artists[:100], start=1):
    print(f"{i}. {artist}")
    artist_list.append(artist)

# Close the browser session
driver.quit()
```

```python
# Removing Duplicates in case of Artists having Multiple Songs on the Top 100
unique_artists = list(set(artist_list))


# Storing the Data in a TXT file
with open('trending_artists.txt', 'w') as file:
    for artist in unique_artists:
        file.write(f"{artist}\n")
```

```python
# Getting the List of Trending Actors & Actresses from IMDB's Starmeter


# Importing relevant packagaes
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
import time

# Initialize the undetected Chrome driver
driver = uc.Chrome()

# Open IMDb STARmeter page
url = "https://www.imdb.com/chart/starmeter/"
driver.get(url)
actor_list = []

# Wait for the page to load
time.sleep(3)  # Adjust this based on internet speed and page load time

# Find all elements containing the actor names
actor_elements = driver.find_elements(By.CLASS_NAME, 'ipc-title__text')

# Extract the text (actor names) from the elements
actors = [element.text for element in actor_elements]

# Print the top 10 actors
print("Trending Actors on IMDb (STARmeter):")
for i, actor in enumerate(actors[2:102], start=1):  # Adjust to get more names if needed
    print(f"{i}. {actor}")
    actor_list.append(actor)

# Close the browser session
driver.quit()
```

```python
# Duplicating any possible Duplicates from the List of Actors Scraped
unique_actors = list(set(actor_list))


# Storing the Data in a TXT file
with open('trending_actors.txt', 'w') as file:
    for actor in unique_actors:
        file.write(f"{actor}\n")
```
